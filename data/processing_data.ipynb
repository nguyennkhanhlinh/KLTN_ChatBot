```python
data = df.copy()
```

```python

data = data.rename(columns={"Giá theo m²": "Giá theo m² (triệu/m²)", "Tổng giá": "Tổng giá (tỷ)", "Diện tích": "Diện tích (m²)"})
data
```

```python
data.describe()
```

```python

sns.boxplot(data=data['Số phòng ngủ'])
plt.show()
```

```python
Q1 = data['Số phòng ngủ'].quantile(0.25)
Q3 = data['Số phòng ngủ'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_IQR = data[(data['Số phòng ngủ'] >= lower_bound) & (data['Số phòng ngủ'] <= upper_bound)]
data_IQR
```

```python
sns.boxplot(data=data['Số phòng tắm, vệ sinh'])
plt.show()
```

```python
Q1 = data['Số phòng tắm, vệ sinh'].quantile(0.25)
Q3 = data['Số phòng tắm, vệ sinh'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_IQR = data[(data['Số phòng tắm, vệ sinh'] >= lower_bound) & (data['Số phòng tắm, vệ sinh'] <= upper_bound)]
data_IQR
```
